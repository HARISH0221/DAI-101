```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv('Cricket_data.csv')
```

```python
data.info()
```

```python
data.head(10)
```

No of Matches

```python
season_count = data['season'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
season_count.plot(kind='bar', color='skyblue')
plt.title('Number of Matches Played in Each Season')
plt.xlabel('Season')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

Selection of BAT And Bowl based on Toss¶

```python
decision_count = data['decision'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(decision_count, labels=decision_count.index, autopct='%1.1f%%', startangle=90)
plt.title('Decision After Winning Toss')
plt.axis('equal')
plt.show()
```

Home Team vs Away Team Scatter Plot

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['home_runs'], data['away_runs'], color='purple', alpha=0.5)
plt.title('Home Team Runs vs. Away Team Runs')
plt.xlabel('Home Team Runs')
plt.ylabel('Away Team Runs')
plt.show()
```

Home Team Boundaries Distribution

```python
plt.figure(figsize=(8, 6))
plt.boxplot(data['home_boundaries'].dropna(), vert=False)
plt.title('Distribution of Home Team Boundaries')
plt.xlabel('Number of Boundaries')
plt.show()
```

Match Duration¶

```python
plt.figure(figsize=(8, 6))
plt.hist(data['match_days'].dropna(), bins=20, color='orange', edgecolor='black')
plt.title('Match Duration')
plt.xlabel('Duration (Days)')
plt.ylabel('Frequency')
plt.show()
```

Key Player Performance

```python
player_performance = data[['home_key_batsman', 'home_key_bowler', 'home_runs', 'home_wickets']]

player_performance = player_performance.dropna()

plt.figure(figsize=(8, 6))
plt.scatter(player_performance['home_runs'], player_performance['home_wickets'], color='red')
plt.title('Player Performance Analysis')
plt.xlabel('Runs Scored')
plt.ylabel('Wickets Taken')
plt.show()
```

Scores per Venue

```python
venue_avg_scores = data.groupby('venue_name')['home_runs'].mean()

sorted_venues = venue_avg_scores.sort_values(ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(sorted_venues.index, sorted_venues.values, color='green')
plt.title('Venue Analysis')
plt.xlabel('Venue')
plt.ylabel('Average Scores')
plt.xticks(rotation=90)
plt.show()
```

Seasonal Trend Analysis

```python
data['year'] = pd.to_datetime(data['start_date']).dt.year

season_avg_scores = data.groupby('year')['home_runs'].mean()

plt.figure(figsize=(8, 6))
plt.plot(season_avg_scores.index, season_avg_scores.values, marker='o', color='purple')
plt.title('Seasonal Trend Analysis')
plt.xlabel('Year')
plt.ylabel('Average Scores')
plt.xticks(rotation=90)
plt.show()
```

Run Distribution

```python
runs = data['home_runs'].dropna().astype(int)

plt.figure(figsize=(8, 6))
plt.hist(runs, bins=20, color='orange', edgecolor='black')
plt.title('Run Distribution')
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.show()
```

Boundaries¶

```python
boundary_data = data[['home_boundaries', 'away_boundaries']]

cumulative_boundaries = boundary_data.cumsum()

plt.figure(figsize=(10, 6))
plt.stackplot(data.index, cumulative_boundaries['home_boundaries'], cumulative_boundaries['away_boundaries'], labels=['Home Team', 'Away Team'], colors=['green', 'blue'])
plt.title('Boundary Analysis')
plt.xlabel('Matches')
plt.ylabel('Cumulative Boundaries')
plt.legend()
plt.show()
```

Most Player of Match

```python
pom_frequency = data['pom'].value_counts().head(20)

plt.figure(figsize=(10, 6))
plt.barh(pom_frequency.index, pom_frequency.values, color='purple')
plt.title('Player of the Match Analysis')
plt.xlabel('Frequency')
plt.ylabel('Player')
plt.show()
```

Umpire

```python
umpire_data = pd.concat([data['umpire1'], data['umpire2']])

umpire_frequency = umpire_data.value_counts().head(10)

plt.figure(figsize=(10, 6))
plt.barh(umpire_frequency.index, umpire_frequency.values, color='gold')
plt.title('Umpire Analysis')
plt.xlabel('Frequency')
plt.ylabel('Umpire')
plt.show()
```

Season Win Percentage

```python
season_win_percentage = data.groupby('season')['winner'].apply(lambda x: x.value_counts(normalize=True)).unstack()

plt.figure(figsize=(12, 6))
for team in season_win_percentage.columns:
    plt.plot(season_win_percentage.index, season_win_percentage[team], marker='o', label=team)
plt.title('Season-wise Win Percentage')
plt.xlabel('Season')
plt.ylabel('Win Percentage')
plt.legend()
plt.show()
```

Batsman by Boundaries¶

```python
top_batsmen_boundaries = data.groupby('home_key_batsman')['home_boundaries'].sum().nlargest(15)

plt.figure(figsize=(10, 6))
plt.barh(top_batsmen_boundaries.index, top_batsmen_boundaries.values, color='blue')
plt.title('Top Batsmen by Boundaries')
plt.xlabel('Total Boundaries')
plt.ylabel('Batsman')
plt.show()
```

Away Boundaries

```python
top_batsmen_boundaries = data.groupby('away_key_batsman')['away_boundaries'].sum().nlargest(15)

plt.figure(figsize=(10, 6))
plt.barh(top_batsmen_boundaries.index, top_batsmen_boundaries.values, color='blue')
plt.title('Top Batsmen by Boundaries')
plt.xlabel('Total Boundaries')
plt.ylabel('Batsman')
plt.show()
```

Home bowlers

```python
top_batsmen_boundaries = data.groupby('home_key_bowler')['home_wickets'].sum().nlargest(15)

plt.figure(figsize=(10, 6))
plt.barh(top_batsmen_boundaries.index, top_batsmen_boundaries.values, color='blue')
plt.title('Top Bowlers by Wickets')
plt.xlabel('Total Wickets')
plt.ylabel('Bowler')
plt.show()
```

Away bowlers

```python
top_batsmen_boundaries = data.groupby('away_key_bowler')['away_wickets'].sum().nlargest(15)

plt.figure(figsize=(10, 6))
plt.barh(top_batsmen_boundaries.index, top_batsmen_boundaries.values, color='blue')
plt.title('Top Bowlers by Wickets')
plt.xlabel('Total Wickets')
plt.ylabel('Bowler')
plt.show()
```

Super Over

```python
super_over_matches = data['super_over'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(super_over_matches, labels=super_over_matches.index, autopct='%1.1f%%', colors=['#66b3ff', '#ff9999'])
plt.title('Distribution of Super Over Matches')
plt.show()
```

Team Performance by Toss

```python
team_performance = data.groupby(['home_team', 'decision']).size().unstack()

plt.figure(figsize=(12, 6))
team_performance.plot(kind='bar', stacked=True)
plt.title('Team Performance by Toss Decision')
plt.xlabel('Team')
plt.ylabel('Match Count')
plt.legend(title='Toss Decision')
plt.show()
```

Run Rate Comparison

```python
data['home_run_rate'] = data['home_runs'] / data['home_overs']
data['away_run_rate'] = data['away_runs'] / data['away_overs']

plt.figure(figsize=(8, 6))
data[['home_run_rate', 'away_run_rate']].plot(kind='box')
plt.title('Run Rate Comparison')
plt.ylabel('Run Rate')
plt.xticks([1, 2], ['Home', 'Away'])
plt.show()
```

Match Win Percentage Heat Map¶

```python
win_percentage = data.groupby(['season', 'home_team']).apply(lambda x: (x['home_team'] == x['winner']).mean()).unstack()

plt.figure(figsize=(12, 8))
sns.heatmap(win_percentage, cmap='YlGnBu', annot=True, fmt='.2f', cbar=True)
plt.title('Match Win Percentage by Team and Season')
plt.xlabel('Home Team')
plt.ylabel('Season')
plt.show()
```

```python
win_percentage = data.groupby(['season', 'away_team']).apply(lambda x: (x['home_team'] == x['winner']).mean()).unstack()

plt.figure(figsize=(12, 8))
sns.heatmap(win_percentage, cmap='YlGnBu', annot=True, fmt='.2f', cbar=True)
plt.title('Match Win Percentage by Team and Season')
plt.xlabel('Home Team')
plt.ylabel('Season')
plt.show()
```
